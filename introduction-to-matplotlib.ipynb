 # Introduction to matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.plot();
```

```python
plt.plot()
plt.show()
```

```python
x=[1,2,3,4]
y=[11,22,33,44]
plt.plot(x,y);
```

```python
#method 1
fig=plt.figure()
ax=fig.add_subplot()
plt.show()
```

```python
# method 2
fig=plt.figure()
ax=fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show();
```

```python
#method 3
fig,ax=plt.subplots()
ax.plot(x,y)
```

# MatplotLib example workflow

```python
fig,ax=plt.subplots(figsize=(10,10)) #width height
ax.plot(x,y)
ax.set(title='Simple Plot', xlabel='x-axis', ylabel='y-axis')
fig.savefig('images/sample-plot.png')
```

## Making figures with numpy arrays

```python
x=np.linspace(0,10,100)
x[:10]
```

### Line plot

```python
fig,ax=plt.subplots()
ax.plot(x,x**2)
```

### Scatter plot

```python
fig, ax=plt.subplots()
ax.scatter(x,np.exp(x))
```

```python
fig, ax=plt.subplots()
ax.scatter(x,np.sin(x))
```

```python
# make a plot from dictionary
d={'a': 10, 'b': 8, 'c': 12}

fig,ax=plt.subplots()
ax.bar(d.keys(),d.values())
ax.set(title='dictionary plot',
      ylabel='Price');
```

```python
fig,ax=plt.subplots()
ax.barh(list(d.keys()),list(d.values()));
```

### Histograms

```python
x=np.random.randn(1000)
fig,ax=plt.subplots()
ax.hist(x);
```
