# MatplotLib example workflow

```python
fig,ax=plt.subplots(figsize=(10,10)) #width height
ax.plot(x,y)
ax.set(title='Simple Plot', xlabel='x-axis', ylabel='y-axis')
fig.savefig('images/sample-plot.png')
```

## Making figures with numpy arrays

```python
x=np.linspace(0,10,100)
x[:10]
```

### Line plot

```python
fig,ax=plt.subplots()
ax.plot(x,x**2)
```

### Scatter plot

```python
fig, ax=plt.subplots()
ax.scatter(x,np.exp(x))
```

```python
fig, ax=plt.subplots()
ax.scatter(x,np.sin(x))
```

```python
# make a plot from dictionary
d={'a': 10, 'b': 8, 'c': 12}

fig,ax=plt.subplots()
ax.bar(d.keys(),d.values())
ax.set(title='dictionary plot',
      ylabel='Price');
```

```python
fig,ax=plt.subplots()
ax.barh(list(d.keys()),list(d.values()));
```

### Histograms

```python
x=np.random.randn(1000)
fig,ax=plt.subplots()
ax.hist(x);
```

## Subplot options

```python
fix,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))

ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(d.keys(),d.values());
ax4.hist(np.random.randn(100));
```

```python
# method 2

fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,10))

ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(d.keys(),d.values())
ax[1,1].hist(np.random.randn(100));
```

## plotting from pandas dfs

```python
import pandas as pd
```

```python
car_sales=pd.read_csv('car-sales.csv')
car_sales
```

```python
ts=pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2020',
                                                        periods=1000))
ts=ts.cumsum()
ts.plot()
```

```python
car_sales
```

```python
car_sales["Price"]=car_sales["Price"].str.replace('[\$\,\.]','', regex=True)
car_sales
```

```python
car_sales["Price"]=car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Sale Date"]=pd.date_range('1/1/2020', periods=len(car_sales))
car_sales
```

```python
car_sales["Total sales"]=car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
car_sales.plot(x="Sale Date", y="Total sales")
```

```python
car_sales['Price']=car_sales['Price'].astype(int)
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter")
```
